```python
sns.countplot(x='Dropped_Out', data=df)
plt.title("Student Dropout Distribution")
df["Dropped_Out"].value_counts(normalize=True)*100
plt.show()
# Observation:
# The majority of students did not drop out. Approximately 85% of students remained in school, while about 15% dropped out, indicating a class imbalance in the dataset.
```

```python
sns.histplot(df["Age"], bins=10, kde=True)
plt.title("Age Distribution of Students")
plt.show()
# Observation:
# The majority of students are between 15 and 18 years old, with fewer students aged 19. The distribution is skewed slightly towards younger ages.
```

```python
sns.histplot(df["Final_Grade"], kde=True)
plt.title("Final Grade Distribution")
# Observation:
# The final grades are distributed primarily between 10 and 15, with a peak around 12. There are some lower grades, but the overall distribution suggests a concentration in the middle to upper-middle range.
```

```python
sns.boxplot(x="Dropped_Out", y="Final_Grade", data=df)
plt.title("Final Grade vs Dropout")
# Observation:
# Students who dropped out tend to have lower final grades compared to students who did not drop out. The box plot for dropped-out students shows a lower median and a wider spread towards lower grades.
```

```python
sns.countplot(x="Gender", hue="Dropped_Out", data=df)
plt.title("Gender vs Dropout")
plt.show()
# Observation:
# Both genders show a higher number of students who did not drop out compared to those who did. The proportion of dropped-out students appears similar across genders, but there might be a slightly higher number of females who did not drop out.
```

```python
sns.boxplot(x="Dropped_Out", y="Study_Time", data=df)
plt.title("Study Time vs Dropout")
plt.show()
# Observation:
# Students with less study time (e.g., 1 hour) seem to have a higher tendency to drop out. Conversely, students with more study time appear less likely to drop out, suggesting a positive correlation between study time and staying in school.
```

```python
sns.boxplot(x="Dropped_Out", y="Number_of_Failures", data=df)
plt.title("Failures vs Dropout")
plt.show()
# Observation:
# Students with a higher number of failures are significantly more likely to drop out. The median number of failures for dropped-out students is higher than for non-dropped-out students, and the distribution for dropped-out students extends to a higher number of failures.
```

```python
sns.boxplot(x="Dropped_Out", y="Number_of_Absences", data=df)
plt.title("Absences vs Dropout")
plt.show()
# Observation:
# Students who dropped out generally have a higher number of absences. While both groups show some outliers, the median absences for dropped-out students are notably higher, indicating a correlation between increased absences and dropout.
```

```python
sns.countplot(x="Internet_Access", hue="Dropped_Out", data=df)
plt.title("Internet Access vs Dropout")
plt.show()
# Observation:
# Students with internet access have a higher overall count, and while both groups (with and without internet) show dropouts, the proportion of dropped-out students appears higher among those without internet access. This suggests internet access might be a contributing factor to staying in school.
```

```python
sns.countplot(x="Family_Support", hue="Dropped_Out", data=df)
plt.title("Family Support vs Dropout")
# Observation:
# A higher number of students who did not drop out reported having family support. Among students who dropped out, the distribution across different levels of family support seems more varied, but generally, less family support might be associated with a higher likelihood of dropping out.
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
# Observation:
# The heatmap reveals various correlations between numerical features. Notable positive correlations include Grade_1, Grade_2, and Final_Grade. Number_of_Failures shows negative correlations with grades. Weekday and Weekend Alcohol Consumption are positively correlated.
```

```python
sns.boxplot(df["Number_of_Absences"])
plt.title("Outliers in Absences")
# Observation:
# The boxplot for 'Number_of_Absences' clearly indicates the presence of significant outliers, with many students having an unusually high number of absences compared to the majority, which could impact models if not handled.
```

```python
sns.pairplot(
    df[['Study_Time','Number_of_Failures','Number_of_Absences','Final_Grade','Dropped_Out']],
    hue='Dropped_Out'
)
# Observation:
# The pairplot visualizes relationships between selected variables, segmented by 'Dropped_Out' status. It shows distinct patterns, such as lower grades and higher failures/absences being more common among dropped-out students, providing a multi-variate view of these interactions.
```
